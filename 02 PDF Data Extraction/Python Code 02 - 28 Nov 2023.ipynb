# **Setup**

```python
import regex as re
from PyPDF4 import PdfFileReader,PdfFileWriter
from IPython.display import clear_output
# from pdfminer import extract_pages, extract_text
```

# **Data Import**

```python
# Read PDF File
file = PdfFileReader('pdfminer-docs.pdf')
file
```

# **Extract Text Data**

```python
# Get Page No 00
# file.getPage(0)
```

```python
File_Text_Data = ' '

for i in range(file.getNumPages()):
    print('Processing Page No.:',i)
    clear_output(wait=True)

    # Extract Text
    Page_Text_Data = file.getPage(i).extractText()
    print('Text Data Size: ',len(Page_Text_Data))

    # Merge Page Text With File Text
    File_Text_Data += Page_Text_Data
    print('Text Data Size: ',len(File_Text_Data),'\n')
```

```python
File_Text_Data
```

## **Extract Words**

```python
File_Words = re.findall(r'\w+',File_Text_Data)

print('No. Of Words: ',len(File_Words))
File_Words[0:10]
```

## **Cleaned Text**

```python
File_Text_Cleaned = ' '.join(File_Words)

print('Text Data Size: ',len(File_Text_Cleaned))
File_Text_Cleaned
```
